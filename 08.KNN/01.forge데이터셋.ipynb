```python
from sklearn.model_selection import train_test_split
X,y=mglearn.datasets.make_forge()
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
#knn 모형 생성, 이웃의 갯수 3
knn=KNeighborsClassifier(n_neighbors=2) 
```

```python
knn.fit(X_train,y_train) #모형 훈련
```

```python
print(knn.predict(X_test)) #새로운 값 예측
print(y_test)
```

```python
print(knn.score(X_train,y_train)) #학습용 데이터셋의 예측 정확도
print(knn.score(X_test,y_test)) #검증용 데이터셋의 예측 정확도
```

```python
#1행 3열의 서브플롯(가로10,세로3)
fig, axes=plt.subplots(1,3,figsize=(10,3))

#이웃의 갯수가 1,3,9일 경우
for n,ax in zip([1,3,9],axes):
    #knn 모형 생성 및 훈련
    knn=KNeighborsClassifier(n_neighbors=n).fit(X,y) 
    #플로팅
    mglearn.plots.plot_2d_separator(knn,X,fill=True,eps=0.5,ax=ax,alpha=0.4)
    mglearn.discrete_scatter(X[:,0],X[:,1],y,ax=ax) #산점도
    ax.set_title("{0}이웃".format(n))
    ax.set_xlabel("특성0")
    ax.set_ylabel("특성1")
    
axes[0].legend(loc=3)
```
