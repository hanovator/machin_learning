```python
train_cols=df.columns[0:6] #0~5 필드 선택
X=df[train_cols] #독립변수
y=df["Survived"] #종속변수
```

```python
#산점도 행렬 출력
import mglearn
pd.plotting.scatter_matrix(df,c=y,figsize=(15,15),marker="o",cmap=mglearn.cm3)
```

```python
#언더 샘플링
from imblearn.under_sampling import RandomUnderSampler
X_sample,y_sample=RandomUnderSampler(random_state=0).fit_sample(X,y)
X_samp=pd.DataFrame(data=X_sample,columns=train_cols)
y_samp=pd.DataFrame(data=y_sample,columns=["Survived"])
df2=pd.concat([X_samp,y_samp],axis=1)
```

```python
df2["Survived"].value_counts()
```

```python
from sklearn.model_selection import train_test_split

#학습용:검증용 8:2
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10,stratify=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
train_rate=[]
test_rate=[]
neighbors=range(1,11)
for n in neighbors:
    knn=KNeighborsClassifier(n_neighbors=n)
    knn.fit(X_train,y_train)
    train_rate.append(knn.score(X_train,y_train))
    test_rate.append(knn.score(X_test,y_test))
    
plt.rcParams["font.size"]=15
plt.plot(neighbors,train_rate,label="학습용 데이터셋")
plt.plot(neighbors,test_rate,label="검증용 데이터셋")
plt.ylabel("정확도")
plt.xlabel("이웃의 수")
plt.legend()
print(test_rate)
#최적의 k 값 : 5
```

```python
model=KNeighborsClassifier(n_neighbors=5)
model.fit(X_train,y_train)
```

```python
import numpy as np
pred=model.predict(X_test)
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
print(np.mean(pred==y_test))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,pred)
cm
```

```python
df_cm=pd.DataFrame(cm,index=[0,1],columns=[0,1])
df_cm
```

```python
sns.set(font_scale=1.4)
plt.figure(figsize=(9,6))
plt.title("Confusion Matrix")
sns.heatmap(df_cm, annot=True)
```
