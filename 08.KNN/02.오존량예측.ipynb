```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
font_name=font_manager.FontProperties(fname="c:/windows/fonts/malgun.ttf").get_name()
rc("font",family=font_name)

colors=["yellow","red"]
sns.countplot("Result",data=df,palette=colors)
plt.title("분류\n(0:부족, 1:정상)",fontsize=14)
```

```python
cols=["Solar.R","Wind","Temp"]
X=df[cols] #독립변수
y=df["Result"] #종속변수
```

```python
import mglearn
#산점도 행렬 그리기
pd.plotting.scatter_matrix(df,c=y,figsize=(15,15),marker="o", cmap=mglearn.cm3)
```

```python
#언더샘플링
from imblearn.under_sampling import RandomUnderSampler

X_sample,y_sample=RandomUnderSampler(random_state=0).fit_sample(X,y)
X_samp=pd.DataFrame(data=X_sample,columns=cols)
y_samp=pd.DataFrame(data=y_sample,columns=["Result"])
df_samp=pd.concat([X_samp,y_samp],axis=1)
```

```python
df_samp["Result"].value_counts()
```

```python
colors=["yellow","red"]
sns.countplot("Result",data=df_samp,palette=colors)
plt.title("분류\n(0:부족, 1:정상)",fontsize=14)
```

```python
X=X_samp[cols]
y=y_samp["Result"]
```

```python
#학습용:검증용 데이터셋을 8:2로 구분
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.neighbors import KNeighborsClassifier

train_rate=[] #학습용 데이터셋의 정확도를 저정할 리스트
test_rate=[] #검증용 데이터셋의 정확도를 저정할 리스트
neighbors=range(1,11) # k를 1~10 반복 처리

for n in neighbors:
    knn=KNeighborsClassifier(n_neighbors=n) #이웃의 갯수
    knn.fit(X_train,y_train) #모형 훈련
    #리스트에 추가
    train_rate.append(knn.score(X_train,y_train))
    test_rate.append(knn.score(X_test,y_test))
    
plt.rcParams["font.size"]=15
plt.plot(neighbors,train_rate,label="학습용 데이터셋")
plt.plot(neighbors,test_rate,label="검증용 데이터셋")
plt.ylabel("정확도")
plt.xlabel("이웃의 수")
plt.legend()

print(test_rate)
```

```python
#실험을 통해 도출된 최적의 k 값으로 모델 생성
model=KNeighborsClassifier(n_neighbors=8)
model.fit(X_train,y_train) #모델 훈련
```

```python
import numpy as np
pred=model.predict(X_test) #예측값
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
print(np.mean(pred==y_test)) #예측값과 실제값 비교, 평균 계산
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,pred)
cm
```

```python
df_cm=pd.DataFrame(cm,index=[0,1],columns=[0,1])
df_cm
```

```python
sns.set(font_scale=1.4)
plt.figure(figsize=(9,6))
plt.title("Confusion Matrix")
sns.heatmap(df_cm,annot=True)
```
