```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
font_name=font_manager.FontProperties(fname="c:/windows/fonts/malgun.ttf").get_name()
rc("font",family=font_name)

colors=["red","green","blue"]
sns.countplot("Name",data=df,palette=colors)
plt.title("분류\n(0:setosa, 1:versicolor, 2:virginica)",fontsize=14)
```

```python
cols=["SepalLength","SepalWidth","PetalLength","PetalWidth"]
X=df[cols] #독립변수
y=df["Species"] #종속변수
```

```python
#산점도 행렬
import mglearn
pd.plotting.scatter_matrix(df,c=y,figsize=(15,15),marker="o", cmap=mglearn.cm3)
```

```python
from sklearn.model_selection import train_test_split
#8:2로 구분
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=10)
```

```python
from sklearn.neural_network import MLPClassifier
#인공신경망 모형
model=MLPClassifier(random_state=42)
model.fit(X_train,y_train)
```

```python
import numpy as np
pred=model.predict(X_test)
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
print(np.mean(pred==y_test))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,pred)
cm
```

```python
df_cm=pd.DataFrame(cm,index=[0,1,2],columns=[0,1,2])
df_cm
```

```python
sns.set(font_scale=1.4)
plt.figure(figsize=(9,6))
plt.title("Confusion Matrix")
sns.heatmap(df_cm,annot=True)
```
