```python
from sklearn.datasets import make_moons #내장데이터셋

# noise 표준편차, 100개의 샘플
X,y=make_moons(n_samples=100,noise=0.25,random_state=3)
print(X[:5])
print(y[:5])
```

```python
from sklearn.model_selection import train_test_split

#학습용,검증용을 8:2로 구분, stratify=y 클래스를 균등하게 샘플링 (0, 1이 균등하게 분포)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, stratify=y,random_state=42)
```

```python
from sklearn.neural_network import MLPClassifier

#인공신경망 모형을 만들고 훈련
model=MLPClassifier(random_state=0).fit(X_train,y_train)
model
#모델의 hyper parameter(하이퍼 파라미터) 확인
```

```python
print(model.score(X_train,y_train)) #학습용 데이터셋의 정확도
print(model.score(X_test,y_test)) #검증용 데이터셋의 정확도
```

```python
from matplotlib import pyplot as plt
#한글폰트 설정
from matplotlib import font_manager,rc
font_name=font_manager.FontProperties(
    fname="c:/windows/fonts/malgun.ttf").get_name()
rc("font",family=font_name)

#클래스 별로 색상 구분
mglearn.plots.plot_2d_separator(model,X_train,fill=True,alpha=0.3)
#산점도 출력
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
plt.xlabel("특성0")
plt.ylabel("특성1")
```

```python
#입력노드 100개, 은닉노드 10개, 출력노드 1개
#은닉노드 갯수를 100개에서 10개로 줄인 모형
model=MLPClassifier(random_state=0,hidden_layer_sizes=[10])
model.fit(X_train,y_train)
```

```python
print(model.score(X_train,y_train)) #학습용 데이터셋의 정확도
print(model.score(X_test,y_test)) #검증용 데이터셋의 정확도
```
