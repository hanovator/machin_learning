```python
train_cols=df.columns[0:11]
X=df[train_cols] #독립변수
y=df["class"] #종속변수
```

```python
import mglearn
pd.plotting.scatter_matrix(df,c=y,figsize=(15,15),marker="o",
                          cmap=mglearn.cm3)
```

```python
#언더샘플링
from imblearn.under_sampling import RandomUnderSampler
X_sample,y_sample=RandomUnderSampler(random_state=0).fit_sample(X,y)
X_samp=pd.DataFrame(data=X_sample,columns=train_cols)
y_samp=pd.DataFrame(data=y_sample,columns=["class"])
df2=pd.concat([X_samp,y_samp],axis=1)
```

```python
df2["class"].value_counts()
```

```python
X=X_samp[train_cols]
y=y_samp["class"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=10)
```

```python
from sklearn.neural_network import MLPClassifier
#인공신경망 모형 생성
model=MLPClassifier(random_state=0)
#모형 훈련(학습)
model.fit(X_train,y_train)
```

```python
import numpy as np
pred=model.predict(X_test) #예측값
print(model.score(X_train,y_train)) #학습용 데이터셋의 예측 정확도
print(model.score(X_test,y_test)) #검증용 데이터셋의 예측 정확도
print(np.mean(pred==y_test))
```

```python
#스케일링(평균값 0, 표준편차 1)
mean_on_train=X_train.mean(axis=0)
std_on_train=X_train.std(axis=0)
X_train_scaled=(X_train - mean_on_train)/std_on_train
X_test_scaled=(X_test - mean_on_train)/std_on_train
```

```python
#스케일링된 학습용 데이터셋으로 훈련
model=MLPClassifier(random_state=0)
model.fit(X_train_scaled,y_train)
```

```python
print(model.score(X_train_scaled,y_train)) #학습용 데이터셋의 예측 정확도
print(model.score(X_test_scaled,y_test)) #검증용 데이터셋의 예측 정확도
```

```python
#최대 반복 횟수 max_iter를 200에서 300으로 조정
model=MLPClassifier(random_state=0,max_iter=300)
model.fit(X_train_scaled,y_train)
```

```python
print(model.score(X_train_scaled,y_train)) #학습용 데이터셋의 예측 정확도
print(model.score(X_test_scaled,y_test)) #검증용 데이터셋의 예측 정확도
```

```python
#은닉노드 1000개로 설정
model=MLPClassifier(random_state=0,max_iter=500,hidden_layer_sizes=[1000])
model.fit(X_train_scaled,y_train)
```

```python
print(model.score(X_train_scaled,y_train)) #학습용 데이터셋의 예측 정확도
print(model.score(X_test_scaled,y_test)) #검증용 데이터셋의 예측 정확도
```

```python
#은닉층을 2개로 설정
model=MLPClassifier(random_state=0,max_iter=300,hidden_layer_sizes=[100,100])
model.fit(X_train_scaled,y_train)
```

```python
print(model.score(X_train_scaled,y_train)) #학습용 데이터셋의 예측 정확도
print(model.score(X_test_scaled,y_test)) #검증용 데이터셋의 예측 정확도
```

```python
#은닉층을 3개로 설정
model=MLPClassifier(random_state=0,
  max_iter=300,hidden_layer_sizes=[100,100,100])
model.fit(X_train_scaled,y_train)
```

```python
print(model.score(X_train_scaled,y_train)) #학습용 데이터셋의 예측 정확도
print(model.score(X_test_scaled,y_test)) #검증용 데이터셋의 예측 정확도
```

```python
from sklearn.metrics import confusion_matrix
pred=model.predict(X_test_scaled)
cm=confusion_matrix(y_test,pred)
cm
```

```python
df_cm=pd.DataFrame(cm,index=[0,1],columns=[0,1])
df_cm
```

```python
sns.set(font_scale=1.4)
plt.figure(figsize=(9,6))
plt.title("Confusion Matrix")
sns.heatmap(df_cm,annot=True,fmt="d")
```
