```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer

#내장 데이터셋 로딩
cancer=load_breast_cancer()
#print(cancer)
#data 독립변수, target 종속변수, feature_names 변수명
df=pd.DataFrame(np.c_[cancer["data"],cancer["target"]], columns=np.append(cancer["feature_names"],["target"]))
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
%matplotlib inline

font_name=font_manager.FontProperties(fname="c:/windows/fonts/malgun.ttf").get_name()
rc("font",family=font_name)

colors=["yellow","red"]

sns.countplot("target",data=df, palette=colors)
plt.title("분류\n, (0:양성, 1:악성)", fontsize=14)
```

```python
train_cols=df.columns[0:27]
X=df[train_cols]
y=df["target"]
```

```python
#언더샘플링
from imblearn.under_sampling import RandomUnderSampler

X_sample,y_sample=RandomUnderSampler(random_state=0).fit_sample(X,y)
X_samp=pd.DataFrame(data=X_sample, columns=train_cols)
y_samp=pd.DataFrame(data=y_sample, columns=["target"])
df2=pd.concat([X_samp,y_samp],axis=1)
```

```python
df2["target"].value_counts()
```

```python
X=X_samp[train_cols]
y=y_samp["target"]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
#트리 모형 생성
model=DecisionTreeClassifier(random_state=0)
#모형 훈련
model.fit(X_train,y_train)
```

```python
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
```

```python
#최대 질문 단계 4로 가지치기
model=DecisionTreeClassifier(max_depth=4,random_state=0)
model.fit(X_train,y_train)
```

```python
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
```

```python
from sklearn.metrics import confusion_matrix
pred=model.predict(X_test)
cm=confusion_matrix(y_test,pred)
cm
```

```python
df_cm=pd.DataFrame(cm,index=[0,1],columns=[0,1])
df_cm
```

```python
sns.set(font_scale=1.4)
plt.figure(figsize=(9,6))
plt.title("Confusion Matrix")
sns.heatmap(df_cm,annot=True)
```

```python
from sklearn.tree import export_graphviz
import pydotplus
import graphviz
from IPython.display import Image
#트리 모델 생성
dot_data=export_graphviz(model,out_file=None,
    feature_names=train_cols,class_names=["bad","good"],
    filled=True,rounded=True,special_characters=True)
graph=pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
import numpy as np
rc("font",family=font_name)

def plot_tree(model):
    n_features=X.shape[1]
    plt.rcParams["figure.figsize"]=(16,9)
    plt.barh(range(n_features),model.feature_importances_,
             align="center")
    plt.yticks(np.arange(n_features),train_cols)
    plt.xlabel("특성 중요도")
    plt.ylabel("특성")
    plt.ylim(-1,n_features)
    
plot_tree(model)    
```

```python
#랜덤포레스트 트리100개
from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier(n_estimators=100,random_state=10)
forest.fit(X_train, y_train)
print(forest.score(X_train,y_train))
print(forest.score(X_test,y_test))
```

```python
#max_depth를 3으로 조절
forest=RandomForestClassifier(
    n_estimators=100,random_state=10,max_depth=4)
forest.fit(X_train, y_train)
print(forest.score(X_train,y_train))
print(forest.score(X_test,y_test))
```

```python
plot_tree(forest)
```
