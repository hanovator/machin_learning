```python
import seaborn as sns
colors=["red","green","blue"]
sns.countplot("Name",data=df,palette=colors) #품종별 카운트
plt.title("분류\n(0:setosa, 1:versicolor, 2:virginaca)",fontsize=14)
```

```python
cols=["SepalLength","SepalWidth","PetalLength","PetalWidth"]
X=df[cols] #독립변수
y=df["Species"] #종속변수
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
#학습용:검증용 8:2로 구분
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2, random_state=10)
#의사결정나무 모형 생성
model=DecisionTreeClassifier(random_state=0)
#모형 훈련(학습)
model.fit(X_train, y_train)
```

```python
print(model.score(X_train,y_train)) #학습용 데이터셋의 예측 정확도
print(model.score(X_test,y_test)) #검증용 데이터셋의 예측 정확도
```

```python
from sklearn.metrics import confusion_matrix
pred=model.predict(X_test)
cm=confusion_matrix(y_test,pred) #실제값과 예측값 비교
cm
```

```python
df_cm=pd.DataFrame(cm,index=[0,1,2],columns=[0,1,2])
df_cm
```

```python
sns.set(font_scale=1.4)
plt.figure(figsize=(9,6))
plt.title("Confusion Matrix")
sns.heatmap(df_cm,annot=True)
```

```python
from sklearn.tree import export_graphviz
import pydotplus
import graphviz
from IPython.display import Image
import os
os.environ["PATH"] += os.pathsep+"c:/Program Files (x86)/Graphviz2.38/bin"

dot_data=export_graphviz(model,out_file=None, feature_names=cols,
                         class_names=["setosa","versicolor","virginica"],
                         filled=True,rounded=True,special_characters=True)
graph=pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
import numpy as np

rc("font",family=font_name)

def plot_tree(model):
    n_features=X.shape[1] #변수의 갯수
    print(model.feature_importances_)
    plt.barh(range(n_features),model.feature_importances_,align="center")
    plt.yticks(np.arange(n_features),cols)
    plt.xlabel("특성 중요도")
    plt.ylabel("특성")
    plt.ylim(-1,n_features)
    
plot_tree(model)    
# PetalWidth > PetalLength > SepalWidth > SepalLength 
```

```python
from sklearn.ensemble import RandomForestClassifier
#100개의 트리로 구성된 랜덤포레스트 모형
forest=RandomForestClassifier(n_estimators=100, random_state=10)
#모형 학습
forest.fit(X_train, y_train)
#예측 정확도
print(forest.score(X_train,y_train))
print(forest.score(X_test,y_test))
```

```python
# 랜덤포레스트모델.estimators_[트리인덱스]
export_graphviz(forest.estimators_[2],out_file=None, feature_names=cols,
                         class_names=["setosa","versicolor","virginica"],
                         filled=True,rounded=True,special_characters=True)
graph=pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
plot_tree(forest)
# PetalLength > PetalWidth > SepalLength > SepalWidth
```
