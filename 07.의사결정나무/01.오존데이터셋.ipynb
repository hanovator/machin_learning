```python
#언더샘플링
from imblearn.under_sampling import RandomUnderSampler

cols=["Solar.R","Wind","Temp"]
X=df[cols]
y=df["Result"]
X_sample,y_sample=RandomUnderSampler(random_state=0).fit_sample(X,y)
X_samp=pd.DataFrame(data=X_sample,columns=cols)
y_samp=pd.DataFrame(data=y_sample,columns=["Result"])
df_samp=pd.concat([X_samp,y_samp],axis=1)
```

```python
df_samp["Result"].value_counts()
```

```python
colors=["yellow","red"]
sns.countplot("Result",data=df_samp,palette=colors)
plt.title("분류\n(0:부족, 1:정상)",fontsize=14)
```

```python
cols=["Solar.R","Wind","Temp"]
X=X_samp[cols] #독립변수
y=y_samp["Result"] #종속변수
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

#학습용:검증용 8:2로 구분
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
#의사결정나무 모형 생성 random_state=0 : Seed 고정
model=DecisionTreeClassifier(random_state=0)
#모형 훈련(학습)
model.fit(X_train,y_train)
```

```python
#학습용 데이터셋의 예측 정확도
print(model.score(X_train,y_train))
#검증용 데이터셋의 예측 정확도
print(model.score(X_test,y_test))
# 과적합화 문제
```

```python
#가지치기, 질문의 단계를 3으로 제한
model=DecisionTreeClassifier(max_depth=3,random_state=0)
model.fit(X_train,y_train)
```

```python
#학습용 데이터셋의 예측 정확도
print(model.score(X_train,y_train))
#검증용 데이터셋의 예측 정확도
print(model.score(X_test,y_test))
# 과적합화 문제
```

```python
from sklearn.metrics import confusion_matrix

pred=model.predict(X_test)
cm=confusion_matrix(y_test,pred) #실제값과 예측값 비교(혼동행렬)
cm
```

```python
df_cm=pd.DataFrame(cm,index=[0,1],columns=[0,1])
df_cm
```

```python
sns.set(font_scale=1.4)
plt.figure(figsize=(9,6))
plt.title("Confusion Matrix")
sns.heatmap(df_cm,annot=True)
```

```python
from sklearn.tree import export_graphviz
import pydotplus 
import graphviz
from IPython.display import Image

#graphviz 실행파일의 경로를 환경변수 패스에 추가
import os
os.environ["PATH"] += os.pathsep+"c:/Program Files (x86)/Graphviz2.38/bin"

dot_data=export_graphviz(model,out_file=None,
    feature_names=cols, class_names=["bad","good"],
    filled=True, rounded=True, special_characters=True)
graph=pydotplus.graph_from_dot_data(dot_data)    
Image(graph.create_png())
```

```python
import numpy as np
from matplotlib import font_manager,rc
font_name=font_manager.FontProperties(
    fname="c:/windows/fonts/malgun.ttf").get_name()
rc("font",family=font_name)

#특성 중요도를 출력하는 함수
def plot_tree(model):
    n_features=X.shape[1]  #독립변수의 갯수
    #print(model.feature_importances_)
    plt.barh(range(n_features), model.feature_importances_, align="center")
    plt.yticks(np.arange(n_features),cols)
    plt.xlabel("특성 중요도")
    plt.ylabel("특성")
    plt.ylim(-1,n_features)
    
#함수 호출    
plot_tree(model)    
#중요한 변수 : Temp > Solar.R > Wind
```

```python
from sklearn.ensemble import RandomForestClassifier
#랜덤포레스트 모델 생성
#n_estimators : 트리갯수
forest=RandomForestClassifier(n_estimators=100,random_state=100,max_depth=3)
forest.fit(X_train, y_train)  #모델 훈련
```

```python
print(forest.score(X_train,y_train)) #학습용 데이터셋의 정확도
print(forest.score(X_test,y_test)) #검증용 데이터셋의 정확도
```

```python
plot_tree(forest) #트리보다 좀더 다양한 시각으로 데이터를 분석
```
