```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer

cancer=load_breast_cancer() #내장된 데이터셋 로딩
print(cancer)
#데이터프레임으로 변환
df=pd.DataFrame(np.c_[cancer["data"], cancer["target"]],
    columns=np.append(cancer["feature_names"],["target"]))
print(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
%matplotlib inline
font_name=font_manager.FontProperties(fname="c:/windows/fonts/malgun.ttf").get_name()
rc("font",family=font_name)

colors=["yellow","red"]
sns.countplot("target",data=df,palette=colors)
plt.title("분류\n(0:양성, 1:악성)",fontsize=14)
```

```python
train_cols=df.columns[0:27]
X=df[train_cols] #독립변수
y=df["target"] #종속변수
```

```python
#언더샘플링
from imblearn.under_sampling import RandomUnderSampler

X_sample,y_sample=RandomUnderSampler(random_state=0).fit_sample(X,y)
X_samp=pd.DataFrame(data=X_sample,columns=train_cols)
y_samp=pd.DataFrame(data=y_sample,columns=["target"])
df2=pd.concat([X_samp,y_samp],axis=1)
```

```python
df2["target"].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
%matplotlib inline
font_name=font_manager.FontProperties(fname="c:/windows/fonts/malgun.ttf").get_name()
rc("font",family=font_name)

colors=["yellow","red"]
sns.countplot("target",data=df2,palette=colors)
plt.title("분류\n(0:양성, 1:악성)",fontsize=14)
```

```python
import statsmodels.api as sm

model=sm.Logit(y,X) #모델 생성
result=model.fit()  #모델 훈련(학습)
result.summary2() #모델의 요약정보 출력
#모델의 설명력
#유의하지 않은 변수들
#변수의 중요도
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

#데이터셋 구분 8:2
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
model=LogisticRegression() #모형 생성
model.fit(X_train,y_train) #모형 훈련(학습)
print(model.score(X_train, y_train)) #학습용 데이터셋의 정확도
print(model.score(X_test, y_test)) #검증용 데이터셋의 정확도
```

```python
from sklearn.metrics import confusion_matrix
pred=model.predict(X_test)
cm=confusion_matrix(y_test, pred) #실제값과 예측값 비교
cm
```

```python
df_cm=pd.DataFrame(cm, index=[0,1],columns=[0,1])
df_cm
```

```python
sns.set(font_scale=1.4)
plt.figure(figsize=(9,6))
plt.title("Confusion Matrix")
sns.heatmap(df_cm, annot=True)
```
