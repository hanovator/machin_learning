```python
train_cols=df.columns[0:29]
X=df[train_cols] #독립변수
y=df["Class"] #종속변수
```

```python
#오버샘플링
from imblearn.over_sampling import SMOTE
X_sample,y_sample=SMOTE(random_state=0).fit_sample(X,y)
X_samp=pd.DataFrame(data=X_sample, columns=train_cols)
y_samp=pd.DataFrame(data=y_sample, columns=["Class"])
df2=pd.concat([X_samp, y_samp], axis=1)
```

```python
df2["Class"].value_counts() #클래스별 샘플 갯수
```

```python
colors=["yellow","red"]
sns.countplot("Class",data=df2, palette=colors)
plt.title("분류\n(0:정상, 1:사기)",fontsize=14)
```

```python
import statsmodels.api as sm
model=sm.Logit(y,X) #로지스틱 회귀 모형
result=model.fit() #모형 훈련(학습)
result.summary2() #요약표
#모델의 설명력 : 6.3%
```

```python
X=df[['Time', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 
      'V9', 'V10',
       'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 
       'V20',
       'V21', 'V22', 'V23', 'V25', 'V26', 'V27']] #독립변수
y=df["Class"] #종속변수
```

```python
import statsmodels.api as sm
model=sm.Logit(y,X) #로지스틱 회귀 모형
result=model.fit() #모형 훈련(학습)
result.summary2() #요약표
#모델의 설명력 : 6.3%
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
#학습용:검증용 = 8:2
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,
                                              random_state=0)
model=LogisticRegression() #로지스틱 회귀 모형
model.fit(X_train,y_train) #모형 훈련
```

```python
print(model.score(X_train, y_train)) #학습용 예측 정확도
print(model.score(X_test, y_test)) #검증용 예측 정확도
```

```python
from sklearn.metrics import confusion_matrix
pred=model.predict(X_test)
cm=confusion_matrix(y_test,pred) #실제값과 예측값 비교
cm
```

```python
df_cm=pd.DataFrame(cm,index=[0,1],columns=[0,1])
df_cm
```

```python
import seaborn as sns

sns.set(font_scale=1.4)
plt.figure(figsize=(9,6))
plt.title("Confusion Matrix")
sns.heatmap(df_cm, annot=True, fmt="d")
```
