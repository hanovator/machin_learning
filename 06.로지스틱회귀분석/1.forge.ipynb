```python
%matplotlib inline
from matplotlib import pyplot as plt
import mglearn #mglearn 패키지를 미리 설치해야 함
#한글 폰트 설정
from matplotlib import font_manager, rc
font_name=font_manager.FontProperties(
    fname="c:/windows/fonts/malgun.ttf").get_name()
rc("font",family=font_name)
```

```python
plt.rcParams["figure.dpi"]=300  # dpi 설정
X,y=mglearn.datasets.make_forge() # 임의로 생성한 데이터셋
print("x === ",X[:5])
print("y === ",y[:5])
mglearn.discrete_scatter(X[:,0],X[:,1],y) #산점도
plt.legend(["클래스0","클래스1"],loc=4) #범례 위치 지정
plt.xlabel("첫번째 특성")
plt.ylabel("두번째 특성")
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

#서브플롯 1행 1열, figsize=(가로인치,세로인치)
fig,axes=plt.subplots(1,1,figsize=(5,3))
model=LogisticRegression() #로지스틱 회귀분석 모델 생성
clf=model.fit(X,y) #모델 훈련(학습)

#2개의 클래스 구분선 그리기
mglearn.plots.plot_2d_separator(clf,X,fill=False,eps=0.5,ax=axes,alpha=0.7)
mglearn.discrete_scatter(X[:,0],X[:,1],y,ax=axes)
axes.set_title("로지스틱 회귀분석")
axes.set_xlabel("특성0")
axes.set_ylabel("특성1")
axes.legend() #범례 출력
```
