```python
train_cols=df.columns[1:17] #1~16 필드
X=df[train_cols] #독립변수
y=df["class_type"] #종속변수
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
model=LogisticRegression() #로지스틱 회귀분석 모형 생성
model.fit(X_train,y_train) #모형 훈련(학습)
```

```python
print(model.score(X_train,y_train)) #학습용
print(model.score(X_test,y_test)) #검증용
```

```python
from sklearn.metrics import confusion_matrix

pred=model.predict(X_test)
cm=confusion_matrix(y_test,pred) #실제값과 예측값 비교(혼동행렬,오분류표)
cm
```

```python
df_cm=pd.DataFrame(cm,index=[1,2,3,4,5,6,7],columns=[1,2,3,4,5,6,7])
df_cm
```

```python
sns.set(font_scale=1.4) #seaborn 패키지의 폰트 사이즈 설정
plt.figure(figsize=(9,6)) #가로,세로 인치
plt.title("Confusion Matrix")
sns.heatmap(df_cm,annot=True) #오분류표 출력
```
