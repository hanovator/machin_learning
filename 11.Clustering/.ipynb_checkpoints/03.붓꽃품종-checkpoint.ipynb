```python
from sklearn import datasets
iris=datasets.load_iris()
iris
```

```python
import pandas as pd
X=pd.DataFrame(iris.data) #독립변수
X.columns=["Sepal_Length","Sepal_Width","Petal_Length",
          "Petal_Width"]
y=pd.DataFrame(iris.target) #종속변수(품종)
y.columns=["Targets"]
df=pd.concat([X,y],axis=1)
df.head()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
font_name=font_manager.FontProperties(
    fname="c:/windows/fonts/malgun.ttf").get_name()
rc("font",family=font_name)

plt.figure(figsize=(14,7))
colormap=np.array(["red","lime","black"])

plt.subplot(1,2,1)
plt.scatter(X.Sepal_Length,X.Sepal_Width,c=colormap[y.Targets],s=40)
plt.title("꽃받침")

plt.subplot(1,2,2)
plt.scatter(X.Petal_Length,X.Petal_Width,c=colormap[y.Targets],s=40)
plt.title("꽃잎")
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=3,random_state=0) #군집 갯수 3
model.fit(X) #데이터만 입력하고 답은 알려주지 않음(비지도학습)
```

```python
model.labels_  #모형에서 분류한 레이블
```

```python
plt.figure(figsize=(14,7)) #가로,세로 인치
colormap=np.array(["red","lime","black"]) #품종별 색상

plt.subplot(1,2,1) #1행 2열 중 1번 그래프
plt.scatter(X.Petal_Length,X.Petal_Width,c=colormap[y.Targets],s=40)
plt.title("정답")

plt.subplot(1,2,2) #1행 2열 중 2번 그래프
plt.scatter(X.Petal_Length,X.Petal_Width,c=colormap[model.labels_],s=40)
plt.title("클러스터링 결과")
```

```python
# 0,1,2를 1,0,2로 바꿈
pred=np.choose(model.labels_, [1,0,2]).astype(np.int64)
```

```python
plt.figure(figsize=(14,7)) #가로,세로 인치
colormap=np.array(["red","lime","black"]) #품종별 색상
plt.subplot(1,2,1) #1행 2열 중 1번 그래프
plt.scatter(X.Petal_Length,X.Petal_Width,c=colormap[y.Targets], s=40)
plt.title("정답")

plt.subplot(1,2,2) #1행 2열 중 2번 그래프
plt.scatter(X.Petal_Length,X.Petal_Width, c=colormap[pred],s=40)
plt.title("클러스터링 결과")
```

```python
import mglearn
from matplotlib import pyplot as plt
fig,axes=plt.subplots(1,4,figsize=(20,5)) #1행 4열의 서브플롯
idx=0
for k in [2,3,4,5]:
    kmeans=KMeans(n_clusters=k) #군집의 갯수 2~5
    kmeans.fit(X) 
    lbls=kmeans.labels_ #모형에서 예측한 값
    #산점도 그리기
    mglearn.discrete_scatter(X["Petal_Length"],X["Petal_Width"],
                            lbls,ax=axes[idx])
    idx += 1
```
