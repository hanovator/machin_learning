```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
X,y=make_blobs(random_state=1) # 
print(X[:5])
print(y[:5])
model=KMeans(n_clusters=3) #KMeans  ,   
model.fit(X) # label  ( )
```

```python
lbl=model.labels_ #
print(lbl) 
```

```python
pred=model.predict(X) #
pred
```

```python
print(len(pred==lbl)/len(lbl)) #
```

```python
import sklearn.metrics as sm
sm.accuracy_score(lbl,pred)
```

```python
mglearn.discrete_scatter(X[:,0],X[:,1],lbl,markers="o")
mglearn.discrete_scatter(model.cluster_centers_[:,0],
                        model.cluster_centers_[:,1],
                        [0,1,2],markers="^",markeredgewidth=5)
```

```python
from matplotlib import pyplot as plt
#1 4 
fig,axes=plt.subplots(1,4,figsize=(20,5)) 
idx=0
for k in [2,3,4,5]: #  2~5
    kmeans=KMeans(n_clusters=k) # 
    kmeans.fit(X) #
    lbls=kmeans.labels_ #  
    # 
    mglearn.discrete_scatter(X[:,0],X[:,1],lbls,ax=axes[idx])
    idx += 1
```
