```python
from sklearn.datasets import load_boston

boston=load_boston() #데이터셋 로딩
print(boston.data.shape) #데이터셋의 형태(차원)
print(boston.data[:3, :]) #첫 3행만 출력
```

```python
boston.feature_names #특성의 이름(변수명)
```

```python
import pandas as pd

# boston.data : 독립변수
# boston.target : 종속변수

X=pd.DataFrame(boston.data, columns=boston.feature_names)
y=pd.DataFrame(boston.target, columns=["MEDV"])
df=pd.concat([X,y],axis=1) #데이터프레임을 열 기준으로 연결함
df.tail()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# 과적합화(over fitting) 방지를 위해 학습용:검증용 데이터셋 구분
# random_state : seed(20)와 같은개념. 20은 중요하지 않다(항상 같은결과를 위해)
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=20)

# 다중회귀분석 모델을 만들고 학습용 데이터를 입력하여 모델을 훈련함(기계학습)
model=LinearRegression().fit(X_train, y_train)
```

```python
# 만들어진 모델에 데이터를 입력하여 예측 정확도를 평가함
print("학습용:",model.score(X_train,y_train))
print("검증용:",model.score(X_test,y_test))
```

```python
#검증용 데이터셋을 입력하여 주택가격을 예측
pred=model.predict(X_test)
pred
```

```python
coef=model.coef_  #상관계수(기울기)
intercept=model.intercept_ #절편
print(coef)
print(intercept)
#범죄율, 비소매상업지역면적비율, 일산화질소 농도, 방의 수, 하위계층비율
#흑인비율, 교사비율, 거주지역비율, 찰스강경계 여부 1/0
```

```python
import statsmodels.formula.api as sm

# 데이터셋을 로딩하여 데이터프레임으로 변환
dfX=pd.DataFrame(boston.data, columns=boston.feature_names)#독립변수
dfy=pd.DataFrame(boston.target, columns=["MEDV"])#종속변수

# 사이킷런이 아닌 다른모델의 회귀분석 모델 생성
model_boston2=sm.OLS(dfy, dfX)

# 회귀분석 모델에 데이터를 입력하여 기계학습 실시
result_boston2=model_boston2.fit()

# 생성된 분석 모델의 해석 
print("회귀계수:\n", result_boston2.params) # 기울기(변수의 영향력)
print("R-squared\n", result_boston2.rsquared) #모델의 설명력
print("p-values\n", result_boston2.pvalues) # 각 변수의 유의확률
# 모델의 요약 정보 출력
print(result_boston2.summary())
```
