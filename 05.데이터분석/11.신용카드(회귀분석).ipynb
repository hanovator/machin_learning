```python
train_cols=df.columns[1:29] #1~28 필드
X=df[train_cols] #독립변수
y=df["Amount"] #종속변수(거래금액)
print(X.head())
print(y.head())
```

```python
import statsmodels.formula.api as sm

model=sm.OLS(y, X) #다중회귀분석 모델 생성
result=model.fit() #데이터를 입력하여 기계학습
result.summary() #모델의 요약표
# R-squared : 81.6%의 설명력
# p-values
# 변수의 중요도(coef)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#학습용:검증용=8:2로 나눔
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2, random_state=10)

#다중회귀분석모델을 만들고 학습용 데이터셋을 입력하여 기계학습 실시
model=LinearRegression().fit(X_train, y_train)

#모델의 예측 정확도 확인
print(model.score(X_train, y_train)) #학습용 데이터셋
print(model.score(X_test, y_test)) #검증용 데이터셋
```

```python
#모델에 새로운 값을 입력하여 결과를 예측함
pre=model.predict(X_test)
print(pre) #예측한 값
print(y_test) #답
```
