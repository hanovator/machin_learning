```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
font_name=font_manager.FontProperties(fname="c:/windows/fonts/malgun.ttf").get_name()
rc("font",family=font_name)

colors=["yellow","red"]
sns.countplot("Result",data=df,palette=colors) #카운트플롯
plt.title("분류\n(0:부족, 1:정상)",fontsize=14)
```

```python
cols=["Solar.R","Wind","Temp"]
X=df[cols] #독립변수
y=df["Result"] #종속변수
```

```python
#산점도 행렬
import mglearn
pd.plotting.scatter_matrix(df,c=y,figsize=(15,15),marker="o", cmap=mglearn.cm3)
```

```python
#언더샘플링
from imblearn.under_sampling import RandomUnderSampler
X_sample,y_sample=RandomUnderSampler(random_state=0).fit_sample(X,y)
X_samp=pd.DataFrame(data=X_sample,columns=cols)
y_samp=pd.DataFrame(data=y_sample,columns=["Result"])
df_samp=pd.concat([X_samp,y_samp],axis=1)
```

```python
df_samp["Result"].value_counts()
```

```python
X=X_samp[cols] #독립변수
y=y_samp["Result"] #종속변수
```

```python
from sklearn.model_selection import train_test_split
#학습용:검증용 8:2로 샘플링
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=20)
```

```python
from sklearn.svm import SVC

model=SVC(random_state=20) #svm 모형 생성
model.fit(X_train,y_train) #학습용 데이터셋을 입력하여 모형 훈련
```

```python
import numpy as np

pred=model.predict(X_test) #예측한 값
print(model.score(X_train,y_train)) #학습용 데이터셋의 예측 정확도
print(model.score(X_test,y_test)) #검증용 데이터셋의 예측 정확도
print(np.mean(pred==y_test))
```

```python
plt.rcParams["figure.figsize"]=(16,9) #그래프의 가로,세로 길이
plt.rcParams["font.size"]=14 #그래프의 폰트 설정
plt.boxplot(X_train) #상자수염그림
plt.xlabel("특성 목록")
plt.ylabel("특성 크기")
```

```python
min_on_training=X_train.min(axis=0) #특성별 최소값
range_on_training=(X_train-min_on_training).max(axis=0) #특성별 최대값
# (데이터 - 최소값) / 최대값 => 최소값을 0, 최대값을 1로 조정하는 작업
X_train_scaled=(X_train - min_on_training) / range_on_training
X_test_scaled=(X_test - min_on_training) / range_on_training
print(X_train_scaled.describe())
print(X_test_scaled.describe())
```

```python
model=SVC(random_state=20)
model.fit(X_train_scaled,y_train) #스케일링된 데이터셋으로 모형 훈련
```

```python
print(model.score(X_train_scaled,y_train))
print(model.score(X_test_scaled,y_test))
```

```python
#최적의 C, gamma를 찾는 과정
import numpy as np
train_rate=[]
test_rate=[]

c_values=[0.001, 0.01, 0.1, 1, 10, 100, 1000] # c값이 클수록 잘분류/과적합
g_values=[0.0001, 0.001, 0.01, 0.1] # 값을 증가시키면 각각의 데이터 포인트가 영향력을 행사하는 거리가 짧아진다.

for n in c_values:
    for g in g_values:
        model=SVC(C=n, gamma=g, random_state=20) #svm 모형 생성
        model.fit(X_train_scaled,y_train) #모형 훈련
        train_rate.append([n,g,model.score(X_train_scaled,y_train)])
        test_rate.append([n,g,model.score(X_test_scaled,y_test)])
        
train_arr=np.array(train_rate) #넘파이 배열로 변환
test_arr=np.array(test_rate) 
max_rate=np.max(test_arr[:,2]) #정확도의 최대값
idx=np.where(test_arr == max_rate)[0][0] #가장 높은 정확도의 인덱스를 찾음
print("최적의 C:",test_rate[idx][0])
print("최적의 gamma:",test_rate[idx][1])
print("최적의 정확도:",test_rate[idx][2])

plt.rcParams["font.size"]=15
plt.plot(range(len(train_rate)),train_arr[:,2],label="학습용 데이터셋")
plt.plot(range(len(test_rate)),test_arr[:,2],label="검증용 데이터셋")
plt.ylabel("정확도")
plt.legend()
```

```python
#최적의 C,gamma value로 모형 생성
model=SVC(C=1, gamma=0.1, random_state=20)
model.fit(X_train_scaled,y_train) #스케일링된 데이터셋으로 모형 훈련
```

```python
print(model.score(X_train_scaled,y_train))
print(model.score(X_test_scaled,y_test))
```

```python
from sklearn.metrics import confusion_matrix
pred=model.predict(X_test_scaled) #예측값
cm=confusion_matrix(y_test,pred) #실제값과 예측값 비교
cm
```

```python
df_cm=pd.DataFrame(cm,index=[0,1],columns=[0,1])
df_cm
```

```python
sns.set(font_scale=1.4)
plt.figure(figsize=(9,6))
plt.title("Confusion Matrix")
sns.heatmap(df_cm, annot=True)
```
